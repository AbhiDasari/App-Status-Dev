```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("final+testcase.csv")
```

```python
data.head()
```

```python
week_weight={"Sunday":4,"Monday":1,"Tuesday":1,"Wednesday":1,"Thursday":1,"Friday":3,"Saturday":3}
data["mWeekDay"]=data["WeekDay"].map(week_weight)
data.head()
```

```python
special_weight={"Christmas":6,"Haloween":4,"Independence Day":4,"New Year":5,"Thanksgiving Week":6}
data["mSpecial"]=data["Special"].map(special_weight)
data=data.fillna(0)
data.head()
```

```python
l=len(data)
for i in range(l):
    data.iloc[i,8]=data.iloc[i,10]+data.iloc[i,11]
data.head()
```

```python
import random
func=["login","Register","Promo","Subscribe","Unsubscribe","Tag","Cart Updation","Badging","Custom Badging","End to End"]
for i in range(l):
    data.iloc[i,9]=random.choice(func)
```

```python
data.head(10)
```

```python
defect=["Functional","Environment","Requirement","Data"]
for i in range(l):
    if(data.iloc[i,2]=="Fail"):
        data.iloc[i,10]=random.choice(defect)
```
